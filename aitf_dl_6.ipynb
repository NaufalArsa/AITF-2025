```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/smirzap/AI-dengan-DL/refs/heads/main/data/BankChurners.csv')
```

```python
df
```

```python
df.info()
```

```python
# Drop unused columns
df = df.drop(['CLIENTNUM', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2', 'Total_Amt_Chng_Q4_Q1', 'Total_Ct_Chng_Q4_Q1'], axis=1)
```

```python
# Merubah kolom “Attrition_Flag” yang merupakan label atau status dari masing-masing klien.
df.rename(columns={'Attrition_Flag':'status'}, inplace=True)

# Melihat jenis status dan jumlah dari setiap klien
df['status'].value_counts()
```

# First Case

```python
# Ambil pelanggan-pelanggan dengan status keluar
df_keluar = df[df['status']=='Attrited Customer']

# Ambil pelanggan-pelanggan dengan status setia
df_setia = df[df['status']=='Existing Customer']
```

```python
# Sampling
df_keluar_sampled = df_keluar.sample(400, random_state=2)
```

```python
# Menggabungkan 400 pelanggan keluar yang telah didapat dengan pelanggan setia:
dataset = pd.concat([df_setia, df_keluar_sampled])
```

```python
dataset['status'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import recall_score
```

```python
# Hapus kolom status sehingga yang tersisa hanya data pokok
data_pokok = dataset.drop('status', axis=1)

# Ambil kolom status dari dataset
data_label = dataset['status']
```

```python
data_pokok.info()
```

```python
data_pokok = pd.get_dummies(data_pokok, drop_first=True)
```

```python
data_pokok.info()
```

```python
data_pokok_train, data_pokok_test, data_label_train, data_label_test = train_test_split(data_pokok, data_label, test_size=0.3, random_state=2)
```

```python
# Modeling
model = DecisionTreeClassifier(max_depth=2,random_state=2)
model.fit(data_pokok_train, data_label_train)

# Show results
from sklearn import tree
rules = tree.export_text(model)
print(rules)
```

```python
# Prediction of status for each customer
prediksi = model.predict(data_pokok_test)

# Count for Recall value
recall_score(data_label_test, prediksi, average=None)
```
